```python
source=cv2.VideoCapture(r"C:\Users\HP\Downloads\4625331-hd_1920_1080_25fps.mp4")  #0 webcamera,path

c=0
while True:
    ret,frame=source.read() ##boolean ,numpy array
    if c==200: break
    if ret==True: # or if frame is not None
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        result=detector.detectMultiScale(gray,1.3,4)
        if len(result)>=1:
            x=result[0][0]
            y=result[0][1]
            w=result[0][2]
            h=result[0][3]
            cropped_image=frame[y:y+h,x:x+w]
            #applying argumentation
            flip=cv2.flip(cropped_image,0)
            gray_col=cv2.cvtColor(cropped_image,cv2.COLOR_BGR2GRAY)
            blur=cv2.GaussianBlur(cropped_image,(3,3),3)
            cv2.imwrite(f"C:/Users/HP/OneDrive/Desktop/faceimages/img_{c}.png",cropped_image) #file name,numpy array
            cv2.imwrite(f"C:/Users/HP/OneDrive/Desktop/faceimages/imgflip_{c}.jpg",flip)
            cv2.imwrite(f"C:/Users/HP/OneDrive/Desktop/faceimages/imggray_{c}.jpg",gray_col)
            cv2.imwrite(f"C:/Users/HP/OneDrive/Desktop/faceimages/imgblur_{c}.jpg",blur)
            c+=1
            
            #plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
            #cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),3) #(x1,y1),(x2,y2)
        cv2.imshow("window",frame)
        if cv2.waitKey(5) & 0xFF==ord('q'):
            break
    else:
        break

cv2.destroyAllWindows()
source.release()
```
